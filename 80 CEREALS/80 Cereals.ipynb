#### Cereals with Highest Sugar Content

```python
sugar_high=df_cereal.loc[df_cereal['sugars'].idxmax()]
print(f"Cereals with highest Sugar: {sugar_high['name']}, {sugar_high['sugars']} cal")
```

```python
sort_sugar=df_cereal.sort_values(by='sugars',ascending=False).head(10)

plt.figure(figsize=(8,5))
plt.xticks(rotation=90)

plt.title('Cereals with High Sugar')
sns.barplot(x=sort_sugar["name"],y=sort_sugar["sugars"],data=sort_sugar),
```

> Golden Crisp and Smacks has highest sugar content

#### Cereals with Lowest Sugar Content

```python
sugar_low=df_cereal.loc[df_cereal['sugars'].idxmin()]
print(f"Cereals with Lower Sugar: {sugar_low['name']}, {sugar_low['sugars']} cal")
```

```python
sort_sugar=df_cereal.sort_values(by='sugars',ascending=False).tail(10)

plt.figure(figsize=(8,5))
plt.xticks(rotation=90)

plt.title('Cereals with Low Sugar')
sns.barplot(x=sort_sugar["name"],y=sort_sugar["sugars"],data=sort_sugar)
```

> Quaker Oat Meal and All-Bran with Extra Fiber has lowest sugar

#### Cereals with High Calories

```python
calories_high=df_cereal.loc[df_cereal['calories'].idxmax()]
print(f"Cereals with highest Calories: {calories_high['name']}, {calories_high['calories']} cal")
```

#### Cereals with Lowest Calories

```python
calories_low=df_cereal.loc[df_cereal['calories'].idxmin()]
print(f"Cereals with lowest Calories: {calories_low['name']}, {calories_low['calories']} cal")
```

#### Distribution of Calories

```python
sns.set(style='whitegrid')
plt.figure(figsize=(10,5))
sns.histplot(df_cereal['calories'],kde=True,color="blue")
plt.title("Distribution of Calories")
plt.show()
```

#### Brands with Highest Rating

```python
cereal_sort=df_cereal.sort_values(by='rating',ascending=False).head(15)
plt.figure(figsize=(10,8))
plt.title("Brands with Top Rating")
plt.xticks(rotation=90)
sns.barplot(data=cereal_sort,x=cereal_sort['name'],y=cereal_sort['rating'])
```

> All-Bran with Extra Fiber and Shredded Wheat 'n'Bran has highest rating


#### Cereals which has highest and lowest protein

```python
high_protein=df_cereal[df_cereal['protein']==df_cereal['protein'].max()]
low_protein=df_cereal[df_cereal['protein']==df_cereal['protein'].min()]


print(f"Cereal with the highest protein: {high_protein['name'].values[0]}, Protein: {high_protein['protein'].values[0]}")
print(f"Cereal with the lowest protein: {low_protein['name'].values[0]}, Protein: {low_protein['protein'].values[0]}")
```

#### Cereals which has highest and lowest vitamin

```python
high_vitamin=df_cereal[df_cereal['vitamins']==df_cereal['vitamins'].max()]
low_vitamin=df_cereal[df_cereal['vitamins']==df_cereal['vitamins'].min()]


print(f"Cereal with the highest vitamins: {high_vitamin['name'].values[0]}, Vitamin: {high_vitamin['vitamins'].values[0]}")
print(f"Cereal with the lowest vitamins: {low_vitamin['name'].values[0]}, Vitamin: {low_vitamin['vitamins'].values[0]}")



```

#### Comparing Calories in each manufacturer

```python
mfr_group = df_cereal.groupby('mfr').mean(numeric_only=True)

plt.figure(figsize=(10, 5))
mfr_group['calories'].sort_values().plot.bar(color='purple')
plt.title('Average Calories by Manufacturer')
plt.ylabel('Calories')
plt.xlabel('Manufacturer')
plt.xticks(rotation=45)
plt.show()
```

```python
print(f"Manufacturers like '{mfr_group['calories'].idxmax()}' have higher calorie cereals on average.")
```
